```python
'''
1) Para una aplicación se necesita validar la fecha ingresada por el usuario. El usuario ingresa un entero 
para el día, un entero para el mes y un entero para el año (año completo, 4 dígitos). No hay que validar 
si los datos ingresados son eneros o no, solo si la fecha es válida. Salida: imprimir “Fecha valida” o 
“Fecha invalida”.
NOTA: Está prohibido utilizar la librería estándar datetime.

dia=int(input(“Ingrese el día:”))
mes=int(input(“Ingrese el mes:”))
anio=int(input(“Ingrese el año:”))
#utilizo la variable anio para evitar la ñ en la variable
…continuar el programa.

'''

borde="========================================"

print(borde)
print("=== Ingresar los siguientes parámetros ===")
print(borde)

dia=int(input("Día:"))
mes=int(input("Mes:"))
año=int(input("Año:"))

if dia <= 0 or mes <= 0 or año <= 0:
    print("\n Fecha Inválida!")
    
elif dia > 31 or mes > 12 or año > 9999:    
    print("\n Fecha Inválida!")

else:
    print("\nFecha Válida!")
    print("\nFecha Ingresada:",dia,"/",mes,"/",año)
    
```

```python
'''
2) Para un programa se necesita imprimir por pantalla un listado de las fichas de dominó. Se puede 
utilizar el “0” (cero) para representar las fichas sin puntos. Tener en cuenta que, por ejemplo, “4-5” y 
“5-4” son la misma ficha y no puede estar duplicada en el listado.

'''

listaValores=[]
listaCombinaciones=[]

valorActual=0
valorMaximo=6


while valorActual <= valorMaximo :
    
    for i in range(1):
        
        listaValores.append(valorActual)
        
        nroCombinaciones=0
        
        for j in range( valorMaximo + 1 ): #Limitamos de 0 a 6
            
            combinacion = listaValores[valorActual] ,  nroCombinaciones
            
            listaCombinaciones.append(combinacion)
            
            nroCombinaciones = nroCombinaciones + 1
        
        valorActual = valorActual + 1
  

listaDomino = listaCombinaciones
    
    
print("\n Lista Valores: ",listaValores)
print("\n Lista Combinaciones:\n\n ",listaCombinaciones)
print("\n Lista Domino:\n\n",listaDomino)
```

```python
'''
5) En una fábrica de panificados se tiene un silo para almacenar la harina. La capacidad del silo debe 
ser un parámetro del programa y se puede cambiar en cada ejecución. Un operario le ingresa la cantidad 
de harina requerida para la preparación (ingreso por teclado). El programa debe informar si la 
cantidad de harina del silo suficiente para realizar la preparación. Si la cantidad es suficiente lo 
informa y actualiza la cantidad harina restante en el silo. Si no es suficiente, lo informa y cancela 
la preparación'''

'''
Nuevamente confuso el enunciado, no es lo mismo la capacidad del silo que la cantidad de harina del 
mismo. Son 2 parametros completamente diferentes, voy a considerar ambos junto con la cantidad de 
harina necesaria.

Esta bueno el ejemplo, pero el desarrollo y resolucion del problema es demasiado ficticio a la realidad.
Estamos hablando de un entorno industrial, entonces vamos a considerar que no existen los sistemas de 
control, procesamiento de señales de sensores, actuadores, etc. Que está directamente arraigado al 
sistema de programación de planta.
A su vez que estamos haciendo una pseudo simulación de control y solo hay 3 parametros
de los 500 a considerar, la capacidad maxima del silo, cantidad actual de harina y la cantidad 
de harina necesaria.
'''

capacidadMaximaSilo=70.0#toneladas logico
cantidadActualSilo=60.0

borde="========================================"
linea="-------------------------------------------------------\n"

print(borde)
print("== Ingresar los siguientes parámetros ==")
print(borde)

cambioCapacidadMaximaSilo=str(input("Desea cambiar la Capacidad Máxima del Silo?( Elija SI/NO )"))
cambioCapacidadMaximaSilo = cambioCapacidadMaximaSilo.lower()
if cambioCapacidadMaximaSilo == 'si':
    capacidadMaximaSiloInput=float(input("Capacidad Máxima del Silo:"))
    capacidadMaximaSilo=capacidadMaximaSiloInput

cambioCantidadActualSilo=str(input("Desea cambiar la Cantidad de Harina Actual del Silo?( Elija SI/NO )"))
cambioCantidadActualSilo = cambioCantidadActualSilo.lower()
if cambioCantidadActualSilo == 'si':
    cantidadActualSiloInput=float(input("Cantidad Actual del Silo:"))
    cantidadActualSilo=cantidadActualSiloInput
    
cantidadHarinaRequerida=float(input("Ingrese la cantidad de harina requerida"))


print(linea)
if cantidadHarinaRequerida < cantidadActualSilo:
    
    cantidadActualSilo = cantidadActualSilo - cantidadHarinaRequerida
    
    print(borde)
    print("== CANTIDAD DE HARINA ADMISIBLE ==")
    print(borde)
    
    
elif cantidadHarinaRequerida >= cantidadActualSilo:
    
    print(borde)
    print("== CANTIDAD DE HARINA NO DISPONIBLE ==")
    print(borde)
    for i in (range(6)):
        print("RELEVAMIENTO EN VERIFICACIÓN.. ")
    
elif capacidadMaximaSilo < cantidadActualSilo :
    
    print(borde)
    print("== FALLA ==")
    print(borde)
    for i in (range(6)):
        print("RELEVAMIENTO EN VERIFICACIÓN.. ")
    



print(linea)
print("== CAPACIDAD MÁXIMA DE HARINA DEL SILO ==")
print(capacidadMaximaSilo,"TON")
print("\n== CANTIDAD ACTUAL DE HARINA DEL SILO ==")
print(cantidadActualSilo,"TON")
print("\n== CANTIDAD DE HARINA REQUERIDA DEL SILO ==")
print(cantidadHarinaRequerida,"TON")

print(linea)

    
    
    
    
```
